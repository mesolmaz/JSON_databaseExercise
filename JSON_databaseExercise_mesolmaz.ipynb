```python
json_str=json.load((open('data/world_bank_projects.json')))
```

```python
df_mjtheme = json_normalize(json_str, 'mjtheme_namecode')
```

```python
df_mjtheme.head(10)
```

```python
df_2 = df_mjtheme.name.value_counts().head(10)
```

```python
df_2.plot(kind='barh').set_title('Top 10 Themes')
```

```python
df_mjtheme.name.isnull().sum()
```

```python
df_mjtheme['name'].replace('', np.nan, inplace=True)
```

```python
df_mjtheme.name.isnull().sum()
```

```python
df_mjtheme.head(10)
```

```python
df_mjtheme.dropna(subset=['name'], inplace=True)
```

```python
df_mjtheme.name.isnull().sum()
```

```python
df_mjtheme_10 = df_mjtheme.dropna().name.value_counts().head(10)
df_mjtheme_10
```
